Newton Method

```python
## TODO: Difference between estimate Equity return and praticle Equity return. 
sigma_init = 0.5

def Equity(A, L, r, sigma, T):
    d1 = (ln(A/L) + (r+0.5*sigma**2)*T)/sigma*sqrt(T)
    d2 = d1 - sigma*sqrt(T)
    return A*N(d1) - L*e(-r*T)*N(d2)

def vega(A, L, r, sigma, T):
    d1 = (ln(A/L) + (r + 0.5*sigma**2)*T)/sigma*sqrt(T)
    firm_vega = A*N(d1)*sqrt(T)
    return firm_vega


def newton_call(A, L, r, T, E, iv):
    Max_iter = 10000
    for i in range(0, Max_iter):
        iv -= ((Equity(A, L, r, iv, T)-E)/vega(A, L, r, iv, T))*0.0001
  ## while abs(Equity(A, L, r, sigma_init, T)-E)/E > 0.1:
  ##   if Equity(A, L, r, sigma_init, T) > E:
  ##     sigma_init -= 0.0001
  ##   else:
  ##     sigma_init += 0.0001
    return iv
```

$E(V_1) = V(1+g)$
